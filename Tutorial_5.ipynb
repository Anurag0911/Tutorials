#### Shallow copy 



M1 copy.copy()

```python
import copy
A = ["apple", 10, 1.2]
B = copy.copy(A)
print(B)
```

```python
A[0] = "orange"
print(A)
print(B)
```

M2 list.copy()

```python
A = ["apple", 10, 1.2]
B = A.copy()
print(B)
```

```python
A[0] = "orange"
print(A)
print(B)
```

M3 Using Slicing

```python
A = ["apple", 10, 1.2]
B = A[:]
A[0] = "banana"
print(A)
print(B)
print(id(A))
print(id(B))
```

The Problem

```python
A = ["apple", 10, 1.2, [1,2,3,4]]
B = A[:]
A[0] = "banana"
A[3][0] = 100
print(A)
print(B)
print(id(A))
print(id(B))
```

Deep copy 


```python
A = ["apple", 10, 1.2, [1,2,3,4]]
B = copy.deepcopy(A)
A[0] = "banana"
A[3][0] = 100
print(A)
print(B)
```
